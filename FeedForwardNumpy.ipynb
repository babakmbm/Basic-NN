```python
import numpy as np
import matplotlib.pyplot as plt

#random data 500 samples and 2 classes
Nclass = 500

#generating 3 gausian clouds 
X1 = np.random.randn(Nclass, 2) + np.array([0,-2])
X2 = np.random.randn(Nclass, 2) + np.array([2,-2])
X3 = np.random.randn(Nclass, 2) + np.array([-2,2])
X = np.vstack([X1,X2,X3])
```

```python
# randomly initialize weights
D = 2 # dimensionality of input
M = 3 # hidden layer size
K = 3 # number of classes
```

```python
#initializing the weighhts and bias
W1 = np.random.randn(D,M) # weights between input and hidden layer
b1 = np.random.randn(M) # bias between input and hidden layer
W2 = np.random.randn(M,K) # weights between and hidden layer and output
b2 = np.random.randn(K) # bias between and hidden layer and output
```

```python
# Sigmoid function
def sigmoid(a):
    return 1 / (1+np.exp(-a))
# we use sigmoid non linearity in the hidden layer and
# softmax for the second layer 
def forward(X, W1, b1, W2, b2):
    Z = sigmoid(X.dot(W1) + b1) # sigmoid
    # Z = np.tanh(X.dot(W1) + b1) # tanh
    # Z = np.maximum(X.dot(W1) + b1, 0) # relu
    A = Z.dot(W2) + b2 
    expA = np.exp(A)
    Y = expA / expA.sum(axis = 1, keepdims=True)
    return Y    
```

```python
def classification_rate(Y,P):
    n_correct = 0
    n_total = 0
    for i in range(len(Y)):
        n_total +=1
        if Y[i] == P[i]:
            n_correct +=1
    return float(n_correct)/n_total
```

```python
P_Y_Given_X = forward(X, W1, b1, W2, b2)
P = np.argmax(P_Y_Given_X, axis = 1)
assert(len(P) == len(Y))

print("Classification rate for random weights:", classification_rate(Y,P))
```
