```python
import numpy as np

# random output Vector
a = np.random.randn(5)
print (a)
```

```python
expa  = np.exp(a)
print (expa)
```

```python
answer = expa / expa.sum()
print (answer)
```

```python
answer.sum()
```

```python
# Random output Matrix A

A = np.random.randn(100, 5)
print(A)
```

```python
expA = np.exp(A)
print(expA)
```

```python
answer  = expA / expA.sum(axis = 1, keepdims = True)
print(answer)
```

```python
answer.sum(axis=1, dtype = float)
```
